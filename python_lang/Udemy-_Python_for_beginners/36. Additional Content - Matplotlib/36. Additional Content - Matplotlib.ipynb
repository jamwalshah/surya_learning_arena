## 319. Generate Line Plot

- `matplotlib.pyplot.ylabel(ylabel: 'str', fontdict: 'dict[str, Any] | None' = None, labelpad: 'float | None' = None, *, loc: "Literal['bottom', 'center', 'top'] | None" = None, **kwargs, )`
    - Set the label for the y-axis.

    - returns 'tuple[np.ndarray | list[np.ndarray], np.ndarray, BarContainer Polygon | list[BarContainer | Polygon]]'
    - Compute and plot a histogram.

    - If the data has already been binned and counted, use `~.bar` or `~.stairs` to plot the distribution::

## 325. Plotting Sub Plots
- So far, we have created a single plot within a given figure
- A figure can have multiple sub-plots as well, and `matplotlib` supports that
- `matplotlib.pyplot.subplots()` method provides a way to plot multiple plots on a single figure
-
- `matplotlib.pyplot.subplots(nrows: 'int' = 1, ncols: 'int' = 1, *, sharex: "bool | Literal['none', 'all', 'row', 'col']" = False, sharey: "bool | Literal['none', 'all', 'row', 'col']" = False, squeeze: 'bool' = True, width_ratios: 'Sequence[float] | None' = None, height_ratios: 'Sequence[float] | None' = None, subplot_kw: 'dict[str, Any] | None' = None, gridspec_kw: 'dict[str, Any] | None' = None, **fig_kw, )`
    - given the `number of rows & cols`, it returns a tuple of `figure`, `axes`
    - returns `(fig, ax)`, giving a single figure `fig` with an array of axes `ax`
    - `fig` :
        - the `matplotlib.pyplot.figure` object to be used as a container for all the subplots
    - `ax` :
        - axes object if there is only one plot, or an array of axes
        - axes object if there are multiple plots, as specified by the `nrows` and `ncols`.
    - `figure` is the container which can hold a single plot or multiple plots
    - `axes` define how you can create multiple sub-plots within the figure
    - `axes[0]` represents the first column
    - Create a figure and a set of subplots.
    - This utility wrapper makes it convenient to create common layouts of subplots, including the enclosing figure object, in a single call.
- `matplotlib.figure.Figure.suptitle(t, **kwargs)`
    - Add a centered suptitle to the figure.


````python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(nrows=1, ncols=2)
print("DEBUG:", type(fig)) # debug
print("DEBUG:", type(ax)) # debug

xpoints = np.array([2020, 2021, 2022, 2023])
ypoints = np.array([6.6, 8.7, 7.0, 6.8])
ax[0].plot(xpoints, ypoints, color="green") # plotting 1st subplot

xpoints = np.array([2020, 2021, 2022, 2023])
ypoints = np.array([-4.5, 3.7, 4.0, 9.8])
ax[1].plot(xpoints, ypoints, color='red') # plotting 2nd subplot

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(nrows=1, ncols=2)
fig.suptitle('Figure with multiple sub plots')
print("DEBUG:", type(fig)) # debug
print("DEBUG:", type(ax)) # debug

xpoints = np.array([2020, 2021, 2022, 2023])
ypoints = np.array([6.6, 8.7, 7.0, 6.8])
ax[0].plot(xpoints, ypoints, color="green") # plotting 1st subplot

xpoints = np.array([2020, 2021, 2022, 2023])
ypoints = np.array([-4.5, 3.7, 4.0, 9.8])
ax[1].plot(xpoints, ypoints, color='red') # plotting 2nd subplot

plt.show()
````
