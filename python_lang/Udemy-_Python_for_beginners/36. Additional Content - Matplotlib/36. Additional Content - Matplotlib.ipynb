## 319. Generate Line Plot
- You'll learn how easy it is to generate graph using `matplotlib` library
- The program keeps running as long as the figure is being rendered, you can close the figure to stop the program
- `matplotlib.pyplot.plot(*args: 'float | ArrayLike | str', scalex: 'bool' = True, scaley: 'bool' = True, data=None, **kwargs,)`
    - Plot y versus x as lines and/or markers.
- `matplotlib.pyplot.show(close=None, block=None)`
    - Display all open figures.
- `matplotlib.pyplot.title(label: 'str', fontdict: 'dict[str, Any] | None' = None, loc: "Literal['left', 'center', 'right'] | None" = None, pad: 'float | None' = None, *, y: 'float | None' = None, **kwargs, )`
    - Set a title for the Axes.
- `matplotlib.pyplot.xlabel(xlabel: 'str', fontdict: 'dict[str, Any] | None' = None, labelpad: 'float | None' = None, *, loc: "Literal['left', 'center', 'right'] | None" = None, **kwargs,)`
    - Set the label for the x-axis.
- `matplotlib.pyplot.ylabel(ylabel: 'str', fontdict: 'dict[str, Any] | None' = None, labelpad: 'float | None' = None, *, loc: "Literal['bottom', 'center', 'top'] | None" = None, **kwargs, )`
    - Set the label for the y-axis.





````python
# matplotlibdemo
# matplot_demo.py
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([2020, 2021, 2022, 2023])
ypoints = np.array([6.6, 8.7, 7.0, 6.8])

plt.plot(xpoints, ypoints) # generates a graph
plt.show() # shows the generated graph
````

````python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([2020, 2021, 2022, 2023])
ypoints = np.array([6.6, 8.7, 7.0, 6.8])

plt.title('Indias GDP by Year') # title for the figure
plt.xlabel('Year') # label for x-axis
plt.ylabel('GDP') # label for y-axis
plt.plot(xpoints, ypoints) # generates a graph
plt.show() # shows the generated graph
````

## 320. Add Color and Grid
- You'll learn how to add color and control a few more things in out line graph
- Area in the figure is called grid, and we can control it
- `matplotlib.pyplot.grid(visible: 'bool | None' = None, which: "Literal['major', 'minor', 'both']" = 'major', axis: "Literal['both', 'x', 'y']" = 'both', **kwargs, )`
    - Configure the grid lines.

````python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([2020, 2021, 2022, 2023])
ypoints = np.array([6.6, 8.7, 7.0, 6.8])

plt.title('Indias GDP by Year') # title for the figure
plt.xlabel('Year') # label for x-axis
plt.ylabel('GDP') # label for y-axis
plt.plot(xpoints, ypoints, color='red') # generates a graph with red color
plt.show() # shows the generated graph
````

````python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([2020, 2021, 2022, 2023])
ypoints = np.array([6.6, 8.7, 7.0, 6.8])

plt.title('Indias GDP by Year') # title for the figure
plt.xlabel('Year') # label for x-axis
plt.ylabel('GDP') # label for y-axis
plt.grid(color='brown', linestyle='--', linewidth=1, axis='y') # creates grid on y-axis only
plt.plot(xpoints, ypoints, color='red') # generates a graph with red color
plt.show() # shows the generated graph
````

````python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([2020, 2021, 2022, 2023])
ypoints = np.array([6.6, 8.7, 7.0, 6.8])

plt.title('Indias GDP by Year') # title for the figure
plt.xlabel('Year') # label for x-axis
plt.ylabel('GDP') # label for y-axis
plt.grid(color='brown', linestyle='--', linewidth=1) # creates grid on both axes
plt.plot(xpoints, ypoints, color='red') # generates a graph with red color
plt.show() # shows the generated graph
````
